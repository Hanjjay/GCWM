### 문제를 해결할 때의 마음가짐

 - 정확하게 문제를 정의할 것
 
 - 작은 것부터 하나하나 확인
 
 - 전체를 바라보면 복합적인 성격을 가지는 문제는 해결할 수 없다
 
 - 전체를 바라보는 것으로 인한 잘못된 문제 해결의 예시
 
 > 스마트폰에 이어폰을 연결하여 노래를 듣고 있었는데 노래가 나오지 않는다
 > 
 > 그래서 스마트폰과 이어폰을 버리고 새로 아이폰11을 구매했다
 >
 > 그러자 노래가 나오지 않는 문제를 해결하였다
 
 - 위에 처럼 하면 비용과 시간이 많이 들어가게 된다 --> 아이폰11은 비싸 + 아이폰11 사러 어딘가에 가야해 (또는 인터넷을 디벼봐야 해)
 
 - 작은 단위부터 문제가 있는지 확인을 하고 전체를 바라보아야 한다!

### 프로그래밍은 문제 해결을 위한 도구!

 - 어떠한 문제를 한번에 전체로 해결하려고 하지 않고 하나하나 접근할 필요가 있다
 
 - 이것은 문제를 해결하는 프로그래밍이라는 도구를 사용할 때도 생각해야할 부분
 
 - 작은 기능 단위부터 구현하고 사용!
 
 - 심지어 작은 단위로 구성한 코드는 상대적으로 쉬울 것이다

### 함수란 무엇인가?

 - 수학 과목에서 배우는 그것
 
 - 매직 박스? 그것
 
 - 영어로는 function 근데 이 단어는 '기능', '도구'라고도 부르는데 ...
 
 - 작은 문제를 해결하기 위한 미니미(?) 프로그램
 
 - 문제의 범위를 잘 설정하는 것이 좋음
 
 - 예시
 
 > - 양은 냄비로 할 수 있는것? --> 양은 냄비는 국도 끓여먹고, 햅반도 중탕해서 먹고, 라면도 끓여먹고, 만두도 쪄먹고 ...
 >
 > - 냄비 비교
 >
 > 1) 진동 감지 기능이 있는, 불빛이 반짝반짝하는, 5 cm의 냄비
 > 
 > 2) 20cm 크기와 10cm 정도의 손잡이 부분에 열전도가 방지가 잘 되는 냄비
 
 - 끓여먹는다는 문제를 잘 정의 했다면 2) 냄비를 고를 것 --> 문제 정의는 중요함
 
 - 하지만 어두운 밤에 덜컹 거리는 차안에서 차를 끓여 먹는다면 1)이 좋을지도 ...? 
 
 - 하지만 일반 냄비라는 함수와 라이트라는 함수와 진동 방지 패드 등등의 여러 함수를 같이 사용하는 게 나중에 더 좋을 것
 
 - 2)은 저 상황 말고 쓸데가 없어~

### 함수라는 것의 핵심

 - 작은 문제로 쪼개어 하나하나 생각
 
 - 문제의 범위에 대한 정의를 잘해야 함
 
 - 이렇게 만들어진 함수는 잘 만들어진 나중에 또 사용할 수 있는 좋은 도구(function)이 될 것!

### python에서 함수를 사용하는 방법

 - 함수 호출하기: 함수이름()

 - 함수에 데이터를 전달하여 호출하기: 함수이름(데이터)

 - 예시
 
 > 함수를 호출하여 출력된 반환값을 보관하기: 결과 = 함수이름(데이터)
 >
 > 함수를 호출하여 출력된 반환값을 화면에 출력하기: print(함수이름(데이터))

```python
import math

print('수를 입력해 주세요')                  # 함수
number = int(input())                        # 함수

result = round(math.sqrt(number * math.pi))  # 함수
print('계산 결과:', result)                  # 함수
```
