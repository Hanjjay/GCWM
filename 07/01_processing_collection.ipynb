##### 연습문제 7-2 리스트와 사전을 이용해 데이터 나타내기

 - 연습문제 7-1에서 정의한 날씨 정보를 리스트와 사전을 이용해 다시 정의해 보자.

#### 양이 많은 데이터를 쉽게 나타내고 읽는 방법

 - 리스트와 사전을 이용하면 데이터를 원하는데로 구조화할 수 있지만 코드가 길어진다
 
 - 즉, 변수가 길어진다
 
 - 이렇게 길어진 구조를 가지는 변수를 쉽고 깔끔하게 표현하는 방법이 없을까?

```python
# 한 행에 걸쳐 모두 표기한 스타일 ... 1)
books1 = [{'title': '파이썬으로 시작하는 컴퓨터 과학 입문', 'authors': ['존 M. 젤']}, {'title': '파이썬을 활용한 데이터 길들이기', 'authors': ['J. 카질', 'K. 자멀']}, {'title': 'HTTP 완벽 가이드', 'authors': ['D. 고울리', 'B. 토티', 'M. 세이어', 'S. 레디']}]

# 리스트를 여러 행에 걸쳐 작성하기 ... 2)
books2 = [
    {'title': '파이썬으로 시작하는 컴퓨터 과학 입문', 'authors': ['존 M. 젤']},
    {'title': '파이썬을 활용한 데이터 길들이기', 'authors': ['J. 카질', 'K. 자멀']},
    {'title': 'HTTP 완벽 가이드', 'authors': ['D. 고울리', 'B. 토티', 'M. 세이어', 'S. 레디']},
]

# 리스트와 리스트에 포함된 사전을 여러 행으로 나누어 작성하기 ... 3)
books3 = [
    {
        'title': '파이썬으로 시작하는 컴퓨터 과학 입문',
        'authors': ['존 M. 젤']
    },
    {
        'title': '파이썬을 활용한 데이터 길들이기',
        'authors': ['J. 카질', 'K. 자멀']
    },
    {
        'title': 'HTTP 완벽 가이드',
        'authors': ['D. 고울리', 'B. 토티', 'M. 세이어', 'S. 레디']
    },
]
```

 - 각 방법에 대한 설명은 아래와 같다.

 > 1) books1은 한 행에 모든 내용을 넣는 방식이다. 이런 방식은 눈으로 봤을 때는 데이터의 내용을 파악하기가 힘들다.
 >
 > 2) books2는 리스트를 여러 행에 걸쳐 작성했다. 리스트의 각각의 요소는 한 행에 하나씩, 한 단계씩 들여쓰기되어 작성되었다. 다만 요소(사전)의 크기가 크다보니 한 행에 표시되는 코드의 양을 줄이기 위해 한 번씩 개행을 했다.
 >
 > 3) books3은 리스트와 사전을 모두 여러 행에 걸쳐 작성한 것이다. 리스트의 요소를 한 단계씩 개행했고, 사전의 키-값 쌍도 한 단계씩 더 개행하여 나타냈다.
 
 - books2와 books3의 스타일처럼 컬렉션의 중첩 구조를 드러내는 스타일을 추천

 - 이렇게 구조화 작업을 진행하는 경우에 데이터가 어떻게 생겨먹었는지 확인인하는게 매우 중요
 
 - 보통 출력(ex - print 함수)으로 확인을 진행하는데 한눈에 잘들어오지 않음
 
 - 이를 위한 함수가 있으니 ...  pprint() 함수(pretty print를 의미)를 사용!

```python
# 컬렉션을 출력했을 때
print(books3)
```

```python
import pprint  # pprint() 함수가 들어있는 모듈을 임포트
pprint.pprint(books3)
```
