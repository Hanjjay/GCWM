### 장고 뷰 만들기

 - 지난 장에서 만들었던 버그를 잡을 시간이 왔다

 - 뷰(view) 는 애플리케이션의 "로직"을 넣는 곳이다
 
 - 뷰는 이전 장에서 만들었던 모델에서 필요한 정보를 받아와서 템플릿에 전달하는 역할을 한다
 
 - 다음 장에서 템플릿을 만들어 볼 것이다
 
 - 뷰는 Python 시작하기 장에서 했던 것보다 어려워보이지만, 파이썬 함수일 뿐이다

 - 뷰는 `views.py` 파일 안에 있습니다. 우리는 *views* 를 `blog/views.py`파일 안에 추가할 것이다
 
### blog/views.py

 - 이제 이 파일을 열고 안에 있는 내용을 살펴보자
