### COVID 19 map을 위한 urls 생성

 - urls 데이터를 추가해보자
 
 - 먼저 blog의 `urls.py`에 COVID19에 접근할 수 있는 url을 추가하자
 
 - 이는 우리의 main url에 뒷부분에 어떤 글자가 붙는가에 따라 다르게 html 파일을 안내하기 위한 작업이다

````python
from django.urls import path
from . import views

urlpatterns = [
    path('', views.post_list, name='post_list'),
    path('COVID19Korea/', views.COVID19_map_data, name='COVID19_map_data'),
]
````

### COVID 19 map을 위한 urls을 보여주기 위한 view setting

 - 위에서 url을 설정하였으나 blogd의 `views.py`에는 이를 안내하기 위한 방법(함수)이 정의되어 있지 않다
 
 - 이를 정의해주어 위에서 설정한 url 방문 시 html 파일을 안내할 수 있게 함수를 만들자
 
 - 아래의 코드를 `views.py`에 입력하자

````python
from django.shortcuts import render

def post_list(request):
    return render(request, 'blog/post_list.html', {})

def COVID19_map_data(request):
    return render(request, 'blog/COVID19/COVID_19_in_South_Korea_200317.html', {})
````
