 - 로컬 컴퓨터는 개발 및 테스트를 수행하는 곳이 될 것이다

### Git 저장소 만들기

 - git을 조금 더 자세하게 설명하자면 코드 저장소(repository: 줄여서 "repo"라고 합니다)로써 특정한 파일들 집합의 변화를 추적하여 관리한다
 
 - 우리는 `djangoW` 디렉토리를 git으로 추적 관리할 것이다
 
 - `djangoW` 디렉토리에서 아래의 명령어를 실행해보자
 
 > `$ git init  
Initialized empty Git repository in ~/djangogirls/.git/  
$ git config --global user.name "Your Name"  
$ git config --global user.email you@example.com  `

 - git 저장소 초기화는 프로젝트를 시작할 때 딱 한 번만 필요합니다 (username과 email 등록 명령은 계속 적용되기 때문에 이 계정에서는 다시 입력할 필요가 없다) 
 
 - username과 email은 이전에 만들어 두었던 github의 계정과 같은 것으로 하여 등록하자
 
 - git은 이 디렉터리에 모든 파일과 폴더들의 변경 점을 추적할 것인데, 특정 파일들을 무시(ignore)하여 추적하지 않게 할 수 있다
 
 - 이는 `.gitignore`파일을 생성하면 가능하다
 
 - `.gitignore`를 사용하여 특정 파일들은 관리하지 않도록 설정하자
 
 > `*.pyc  
*~  
__pycache__  
myvenv  
db.sqlite3  
/static  
.DS_Store  `

 - `djangogW` 폴더의 맨 위에다 `.gitignore` 라는 파일로 저장하자

 > **Note** 파일명 앞에 마침표로 시작하는 것이 중요하다. 만약 파일을 만드는 것이 어렵다면 (예를 들어 맥은 파인더에서 마침표로 시작하는 파일을 생성하는 걸 선호하지 않는다) 다른 이름으로 저장(Save As) 기능을 사용하자.  

 > **Note** .gitignore 파일에 있는 db.sqlite3이라는 파일은 모든 게시물이 저장된 로컬 데이터베이스이다. PythonAnywhere는 다른 데이터베이스를 사용하기 때문에 저장소에 추가될 필요가 없다. 다른 데이터베이스로는 SQLite로도 사용하지만 보통은 SQLite보다 훨씬 많은 방문자를 보유한 웹사이트일 경우 MySQL을 사용한다. GitHub 저장소에 SQLite 데이터베이스를 제외하고 저장하면, 지금까지 작성한 모든 게시물을 로컬에서만 사용할 수 있으므로 실제 운영하는 프로덕션 환경에서는 다시 새 데이터베이스를 추가해야한다. 로컬 데이터베이스는 데이터가 삭제돼도 괜찮은 테스트 공간으로만 사용하자.
 
 - `git add` 명령을 사용하기 전 변경된 것이 있는지 확인이 필요하다면 'git status' 명령어를 사용하는 것은 좋은 방법이다
 
 > 이는 git extension을 사용하는 경우 더욱 쉽게 확인할 수 있다
 
 - 잘못된 파일을 추가하거나 이러한 상황에서 커밋을 수행한 경우 이를 복구할 수 있다
 
 - `git status` 명령은 미추적/수정된/스테이지된 파일, 브랜치 상태와 그 외 많은 정보를 보여준다
 
 - 아래와 같이 실행하고 확인해볼 수 있다
 
 > `$ git status

No commits yet

        .gitignore  
        blog/

        mysite/

 > ` git add --all .  
$ git commit -m "My Django Girls app, first commit"
