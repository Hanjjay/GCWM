### 객체 생성하기

 - 데이터베이스에 새 글 객체를 저장하는 방법에 대해 알아보자  
 
 - 아래의 명령어를 살펴보자  
 
 > **NOTE** 아래의 명령은 모두 django shell에서 진행된다 
 
 `>>> Post.objects.create(author=me, title='Sample title', text='Test')`
 
 - 이 명령어는 블로그에 개시될 글에 대한 하나의 객체이다  
 
 - 여기에 뭔가 빼먹은 것이 하나있다  
  
 - 바로 me(나)를 빼먹었다  
 
 - 작성자로서 User(사용자) 모델의 인스턴스를 가져와 전달해줘야 한다  
 
 - 어떻게 이렇게 할 수 있을까?  

 - 먼저 User 모델을 불러온다
 
 `>>> from django.contrib.auth.models import User`
 
 - 위처럼 사용자라는 모델을 불러와 보았다
 
 - 데이터 베이스에서 user는 어떤 일을 할까?
 
 - 다시 아래의 명령어를 보자
 
 `User.objects.all()`
 
 - 이를 실행하면 아래와 같은 결과를 볼 수 있다

```python
Image("img/07_img.png")
```

 - `User : eornr`으로부터 우리가 이전에 등록한 admin, 슈퍼유저 사용자를 볼 수 있다
 
 - 이 사용자의 인스턴스(instance)를 가져와 보자
 
 - 아래의 명령어를 통해 이 인스턴스를 가져올 수 있다
 
 `>>> me = User.objects.get(username='eornr')`
 
 > **NOTE** eornr 부분에 superuser로 등록한 이름을 작성해야 합니다
 
 - 이 명령어를 실행함으로써 이전에 없었던 `me`에 사용자라는 모델의 인스턴스를 부여했다
 
 - 이제 위에서 실행할 수 없었던 명령어를 실행하여 작성자가 슈퍼유저(우리, 자신)인 게시물을 만들어 낼 수 있다
 
 `Post.objects.create(author=me, title='Sample title', text='Test')`
 
 - 아래의 그림을 통해 이 과정을 확인해보자

```python
Image("img/08_img.png")
```

### 필터링하기  

 - 쿼리셋의 중요한 기능은 데이터를 필터링하는 것이다  
 
 - 예를 들어, 우리는 ola라는 사용자가 작성한 모든 글을 찾고 싶다고 해보자  
 
 - 이런 경우 `Post.objects.all()`에서 `all`대신, `filter`를 사용한다  
 
 - 쿼리셋 안에 있는 괄호 안에 원하는 조건을 넣어줌으로서 찾을 수 있다  
 
 - 지금 이 경우에는 `작성자(author)`가 `나(me)`인 조건을 넣어야할 것이다  
 
 - 이걸 장고로 표현한다면 `author=me`가 된다  
 
 - 이제 이 조건이 반영된 코드를 확인해보자  
 
 > **NOTE** 여기서 나(슈퍼유저, me)가 작성한 글은 4개인데, 이 코드를 실행하여 4개의 글들을 확인할 수 있다

```python
Image("img/09_img.png")
```

 - 만약 `제목(title)`에 'title'이라는 글자가 들어간 글들만 뽑아내어 확인하고 싶다면 아래와 같이 실행하여 확인할 수 있다
 
 > **NOTE** `title`와 `contains` 사이에 있는 밑줄(`_`)이 2개(`__`)이다. 장고 ORM은 필드 이름("title")과 연산자과 필터("contains")를 밑줄 2개를 사용해 구분한다.

```python
Image("img/10_img.png")
```

 - 게시글 목록 중 `게시일(published_date)`로 과거에 작성한 글들을 필터링하여 목록을 불러올 수도 있다
 
 > **NOTE** 여기서 필자는 과거에 게시물을 작성하고 publish를 수행하여 2개의 데이터를 쉽게 불러왔으나, 방금 이 과정을 빠르게 수행한 사람의 경우 빈 리스트가 출력됨을 볼 수도 있다

```python
Image("img/11_img.png")
```

 - 만약 빈 리스트를 확인했다면 지금까지 작성한 글들이 published가 되지 않았기 떄문이다
 
 - 이를 각 게시물을 published하여 빈 리스트가 아닌 값이 출력되게 만들어보자 
 
 - 먼저 게시(published)하려는 게시물의 인스턴스를 얻어야 한다
 
 `>>> post = Post.objects.get(title="Sample title")`
 
 - 이후 `publish` 메서드를 사용하자
 
 `>>> post.publish()`
 
 - 이제 위에서 빈 리스트가 출력된 `filter` 기능을 다시 사용해보자

```python
Image("img/12_img.png")
```

### 정렬하기

- 쿼리셋은 객체 목록을 정렬할 수 있다

- 이제 `created_date` 필드를 기준으로 정렬을 해보자

- 아래의 명령을 실행하여 확인해보자

`>>> Post.objects.order_by('created_date')`

- 여기서 `-`를 맨앞에 붙여주어 내림차순 정렬도 가능하다

`>>> Post.objects.order_by('-created_date')`

```python
Image("img/13_img.png")
```

### 쿼리셋 연결

 - 이전에 이러저러 사용해보았던 쿼리셋을 연결 (chaining)할 수 있다  
 
 - 이를 통해 복잡해보아는 여러 데이터들을 조건에 따라 다룰 수 있다
 
 `>>> Post.objects.filter(published_date__lte=timezone.now()).order_by('published_date')`
