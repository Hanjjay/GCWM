### 템플릿 태그란?

 - HTML에 여러분은 파이썬 코드를 바로 넣을 수 없다  
 
 - 이는 브라우져가 파이썬 코드를 이해할 수 없기 때문이다  
 
 - 인터넷 브라우저는 HTML만을 알고 있다  
 
 - HTML는 정적(static)이지만, 파이썬은 동적(dynamic)이다  

 - 템플릿 태그는 파이썬을 HTML로 바꿔주어, 빠르고 쉽게 동적인 웹 사이트를 만들 수 있게 도와준다
