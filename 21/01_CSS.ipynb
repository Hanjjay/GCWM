#### CSS는 무엇인가요?

 - CSS(Cascading Style Sheets)는 HTML와 같이 마크업언어(Markup Language)로 작성된 웹사이트의 외관을 꾸미기 위해 사용되는 언어이다  
 
 - 웹사이트에 메이크업을 해보자  

 - 아무것도 없는 상태에서 시작하기 어려울 것이다  
 
 - 개발자들이 만든 오픈 소스 코드를 사용해 만들어보자  
 
 - 이미 있는 바퀴를 다시 발명하는 것은 재미도 없고 시간 낭비이다 (있는 것을 잘쓰자!)  
 
#### 부트스트랩을 사용해봐요!

 - 부트스트랩(Bootstrap)은 유명한 HTML과 CSS프레임워크로 예쁜 웹사이트를 만들 수 있다  
 
 > https://getbootstrap.com/  

 - 트위터 개발자들이 부트스트랩을 만들었고, 전 세계 자원봉사자들이 지속적으로 참여해 발전시키고 있다

#### 나의 첫 번째 CSS 파일!

 - CSS파일을 만들어 나만의 스타일을 가진 웹페이지를 만들어보자
 
 - `static` 디렉토리 안에 `css`라고 새로운 디렉토리를 만들자
 
 - 그리고 `css`디렉토리 안에 `blog.css`라는 파일을 만들자
 
 - 아래와 같은 구조를 가질 것이다
 
```
    djangoW
    └─── blog
         └─── static
              └─── css
                   └─── blog.css
```

 - CSS를 쓸 차례이다
 
 - 코드 에디터에서 `blog/static/css/blog.css` 파일을 열어보자

 - 먼저 제목 색상을 바꾸어보자
 
 - 컴퓨터는 특별한 코드를 사용해 색상을 나타내요. #으로 시작해 알파벳(A-F)와 숫자(0-9) 중 6개를 조합해 헥사코드(hexacode)로 나타낸다
 
 - 이 곳에서 원하는 색상 코드를 찾을 수 있다
 
 > https://www.colorpicker.com/ 
 
 - 이러한 코드 없이 빨간색(red) 또는 녹색(green) 등 미리 명시된 색상을 사용할 수도 있다

 - `blog/static/css/blog.css` 파일에 아래 코드를 추가하자

```
h1 a {
    color: #FCA205;
}
```

 - `h1 a`는 *CSS 셀렉터(Selector)* 라고 한다
 
 - `h1`요소 안에 어떠한 `a`요소를 넣어 스타일을 적용 할 수 있다는 의미이다
 
 - 예를 들어 `<h1><a href="">link</a></h1>` 라면, 오렌지색 `#FCA205`으로 바꿀 수 있다
 
 - 물론, 내가 원하는 다른 색상을 지정할 수도 있다
 
 - CSS 파일에서는 HTML 파일에 있는 각 요소들에 스타일을 정의할 수 있다
 
 - 요소를 식별하는 첫 번째 방법은 요소 이름을 사용하는 것이다
 
 - HTML 섹션에서 태그로 기억할 수 있다
 
 - `a`, `h1`, `body` 와 같은 것은 모두 요소 이름의 예이다
 
 - 또 class 속성이나 id 속성에 의해 요소를 식별한다
 
 - 클래스와 ID는 요소에 직접 부여한 이름이다
 
 - 클래스는 요소 그룹을 정의하고 ID는 특정 요소를 가리킨다
 
 - 아래 코드와 같이 태그 이름은 `a`, 클래스는 `external_link` 또는 ID는 `link_to_wiki_page`로 사용하여 태그를 통해 요소를 식별할 수 있습니다
 
 `<a href="https://en.wikipedia.org/wiki/Django" class="external_link" id="link_to_wiki_page">`
 
 - W3Schools에서 CSS 선택자에 대해 읽어보면 매우 좋다
 
 > https://www.w3schools.com/cssref/css_selectors.asp

 - 이제 여러 선택자들을 추가해보자

.top-menu, .top-menu:hover, .top-menu:visited {

    float: right;
    font-size: 26pt;
    margin-right: 20px;
}
