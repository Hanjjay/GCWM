### 템플릿 확장하기

 - 장고의 또 다른 멋진 기능은 템플릿 확장(template extending) 이다  
 
 - 우리가 만들고 있는 웹사이트 안의 서로 다른 페이지에서 HTML의 일부를 동일하게 재사용 할 수 있다는 뜻이다  
 
 - 이 방법을 사용하면 동일한 정보/레이아웃을 사용하고자 할 때, 모든 파일마다 같은 내용을 반복해서 입력 할 필요가 없게 된다
 
 - 또 뭔가 수정할 부분이 생겼을 때, 각각 모든 파일을 수정할 필요 없이 딱 한번만 수정하면 된다

 - 이 코드는 무엇을 의미일까?
 
 - 우리는 `block`을 만든 것이다
 
 - 템플릿 태그 `{% block %}`으로 HTML 내에 들어갈 수 있는 공간을 만들었다
 
 - `base.html`을 확장해 다른 템플릿에도 가져다 쓸 수 있게 한 것이라고 생각할 수 있다
 
 - 이제 이 `block`이 어떻게 동작하는지 확인해보자
 
 - 이제 `base.html`을 파일을 저장을 하고, `blog/templates/blog/post_list.html` 파일을 다시 열자
 
 - `{% for post in posts %}`부터 `{% endfor %}`까지만 남기고 나머지는 전부 지우자
 
 - 우리는 이 코드를 모든 컨텐츠 블록에 대한 템플릿의 일부로 쓸 것이다
 
 - 이 파일에 블록 (`block`) 태그를 추가해보자

 - 블록 태그가 `base.html`파일의 태그와 일치해야할 것이다
 
 - 또 콘텐츠 블록에 속한 모든 코드를 포함하게 만들어야 할 것이다
 
 - 이를 위해서 `{% block content %}`와 `{% endblock %}`사이에 남아있는 모든 코드를 넣어보자
